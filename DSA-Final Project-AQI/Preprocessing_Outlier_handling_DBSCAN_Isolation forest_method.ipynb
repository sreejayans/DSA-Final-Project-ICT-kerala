```python
# Get a list of columns to fill null values in (exclude 'AQI_Bucket' and 'AQI')
columns_to_fill = data.columns.difference(['AQI_Bucket', 'AQI'])

# Fill null values in the selected columns for each city separately
data[columns_to_fill] = data.groupby('City')[columns_to_fill].apply(lambda group: group.fillna(method='ffill').fillna(method='bfill'))

data.head(10)
```

# **DBSCAN**

```python
from sklearn.cluster import DBSCAN
X = data[["AQI",'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2',
       'O3','AQI','AQI_Bucket']]
# Create a DBSCAN model
dbscan = DBSCAN(eps=3, min_samples=2)

# Fit the model to the data
dbscan.fit(X)

# Identify the core samples (i.e., points that belong to a cluster)
core_samples_mask = np.zeros_like(dbscan.labels_, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True

# Identify the outliers (i.e., points that do not belong to any cluster)
outliers = X[~core_samples_mask]

# Remove the outliers from the data
data_without_outliers = X[core_samples_mask]
data_without_outliers.describe()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

# Create a DBSCAN model with eps=3 and min_samples=2
clustering = DBSCAN(eps=3, min_samples=2).fit(data_without_outliers)

# Get the cluster labels
labels = clustering.labels_

# Calculate the Silhouette Coefficient
score = silhouette_score(data_without_outliers, labels)

# Print the results
print(f'Cluster labels: {labels}')
print(f'Silhouette Coefficient: {score:.2f}')
```
